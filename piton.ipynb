```python
#series fibonnacci:
#la suma de dos elementos define el siguiente
a=0
b=1
while b < 100:
  print(b)
  a=b
  b=a+b  #valor de b si esta actualizado
```

```python
a=0
b=1
while b < 100:
  print(b)
  a, b = b, a+b  # El valor es de b, no esta actualizado
```
