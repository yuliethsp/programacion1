1). Cambie una lista a una tupla

```python
lista = ["Taylor Swift", "Clairo", "Olivia Dean", "Sade", "Beabadoobee"]
tupla = tuple(lista)
print("Artistas de mi lista:", lista)
print("Tipo original:", type(lista))

print("\n--- cambio exitoso --- \n")

print("Artistas de la tupla:", tupla)
print("Nuevo tipo de dato", type(tupla))
```

2). Cambie una lista a un diccionario

```python
lista_panaderia = ['croissant', 'dedito de queso', 'pan de bono', 'almojabana']
diccionario = { i: nombre for i, nombre in enumerate(lista_panaderia, start= 0)}
print("lista original:", lista_panaderia)
print(type(lista_panaderia))

print("\n -- cambio a diccionario --\n")

print("Diccionario:", diccionario)
print(type(diccionario))
```

3). Cambie una lista a un conjunto

```python
lista_colores = ['rosado', 'rojo', 'rosado', 'azul', 'naranja', 'morado', 'azul']
conjunto = set(lista_colores)
print("lista original", lista_colores)
print(type(lista_colores))

print("\n -- cambio a conjunto --\n")

print("conjunto:", conjunto)
print(type(conjunto))
```

4). Cambie una tupla a una lista

```python
tupla = ("uno", "dos", "tres", "cuatro", "cinco") #conteo
lista = list(tupla)
print("Tupla original:", tupla)
print(type(tupla))

print("\n -- cambio a lista --\n")
print("Lista:", lista)
print(type(lista))

lista.append("seis")
print("lista con mas datos:", lista)
```

5). Cambie una tupla a un diccionario

```python
tupla_materias = ("Artes", "Ingles", "Matemáticas", "Quimica")
diccionario_materias = {i: materia for i, materia in enumerate(tupla_materias, start=1)}
print("Tupla original", tupla_materias)
print(type(tupla_materias))

print("\n -- Cambio a diccionario --\n")

print("Diccionario:", diccionario_materias)
print(type(diccionario_materias))
```

6). Cambie una tupla a un conjunto

```python
tupla_flores = ("Lirios", "Hortensias", "Margaritas", "Lirios", "Gerberas", "Hortensias")
conjunto_flores = set(tupla_flores)
print("Tupla oriinal:", tupla_flores)
print(type(tupla_flores))

print("\n -- Cambio a conjunto --\n")
print("Conjunto:", conjunto_flores)
print(type(conjunto_flores))
```

7). Cambie un diccionario a una lista

```python
diccionario_juegos = {"1": "Fornite", "2": "Minecraft", "3": "Call of dutty"}
lista_llaves = list(diccionario_juegos.keys())
lista_valores = list(diccionario_juegos.values())
lista_parejas = list(diccionario_juegos.items())
print("Diccionario original:", diccionario_juegos)
print(type(diccionario_juegos))

print("\n -- Cambio a lista --\n")

print("Lista solo llaves:", lista_llaves)
print("Lista solo valores:", lista_valores)
print("Lisa solo items:", lista_parejas)
```

8). Cambie un diccionario a una tupla

```python
diccionario = {"Usuario": "Daniela", "Edad": 22, "Puntos": 2000}
tupla_llave = tuple(diccionario.keys())
tupla_valores = tuple(diccionario.values())
tupla_completa = tuple(diccionario.items())
print("Diccionario original", diccionario)
print(type(diccionario))

print("\n -- Cambio a tupla -- \n")

print("Tupla solo llaves:", tupla_llave)
print("Tupla solo valores:", tupla_valores)
print("Tupla completa:", tupla_completa)
```

9). Cambie un diccionarioo a un conjunto

```python
diccionario = {"User 1": "Activo", "User 2": "Inactivo", "User 3": "Activo", "User 4": "Pendiente"}
set_llaves = set(diccionario.keys())
set_valores = set(diccionario.values())
print("Diccionario original:", diccionario)
print(type(diccionario))

print("\n -- Cambio a un conjunto -- \n")

print("Conjunto de llaves:", set_llaves)
print("Conjunto de valores:", set_valores)
print(type(set_valores))
```

12). Cambie un conjunto a un diccionario

```python
conjunto_carros = {"Kia", "Mercedes", "Suzuki", "BMW"}
diccionario_carros = {elemento: "Carro" for elemento in conjunto_carros}
print("Conjunto original:", conjunto_carros)
print(type(conjunto_carros))

print("\n -- Cambio a diccionario --\n")

print("Diccionario:", diccionario_carros)
print(type(diccionario_carros))
```
